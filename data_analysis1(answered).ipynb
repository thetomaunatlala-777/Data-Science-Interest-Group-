```python
import matplotlib.pyplot as plt

beginner_class = []
intermediate_class = []

for level in data["Knowledge Level"]:
    if level == "Beginner":
        beginner_class.append(level)
    elif level == "Intermediate":
        intermediate_class.append(level)




counts = [len(beginner_class), len(intermediate_class)]
labels = ["Beginner", "Intermediate"]

plt.bar(labels, counts, color=['orange', 'blue'])
plt.title("Counts of Knowledge Levels")
plt.ylabel("Count")
plt.show()
```

We can see that approximately 56,99% of the sampled individuals have a beginner level of knowledge on child sexual abuse and approximately 43,00% of the sampled individuals have an intermediate level of knowledge on child sexual abuse.
